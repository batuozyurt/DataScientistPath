# Analysis of Correlation
---
- Amacımız elimizdeki veri setinin ısı haritası aracılığıyla korelasyonlarını incelemek ve yüksek korelasyona sahip değişkenlerden bazılarını dışarıda bırakmak.
- **Korelasyon**: değişkenlerin birbirleriyle ilişkisini ifade eden bir istatistiksel ölçümdür. **-1** ile **+1** arasında değerler alır.
    - **-1** ve **+1**'e yaklaştıkca ilişkinin şiddeti artar.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = pd.read_csv("breast_cancer.csv")
df = df.iloc[:, 1:-1] # "id" gibi sonucu anlamsız değişkenlerden kurtulmak için.
df.head()
```

- önceliğimiz sayısal değişkenlere ulaşmak.

```python
num_cols = [col for col in df.columns if df[col].dtype in [int, float]]
```

```python
corr = df[num_cols].corr() # korelasyonu hesapladık.
```

```python
sns.set(rc = {'figure.figsize': (12, 12)})
sns.heatmap(corr, cmap = "RdBu")
plt.show()
```

## Yüksek Korelasyonlu Değişkenlerin Silinmesi
---
- Korelasyonun negatif ya da pozitif olmasıyla ilgilenmiyoruz. Bu yüzden mutlak değerleri alarak pozitif duruma getiriyoruz;

```python
cor_matrix = df.corr().abs()
```

```python
cor_matrix
```

```python
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k = 1).astype(bool))
drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > 0.99)]
cor_matrix[drop_list]
```

```python
df.drop(drop_list, axis = 1)
```

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    corr = dataframe.corr()
    cor_matrix = corr.abs()
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={'figure.figsize': (15, 15)})
        sns.heatmap(corr, cmap="RdBu")
        plt.show()
    return drop_list
```

```python
high_correlated_cols(df)
drop_list = high_correlated_cols(df, plot=True)
df.drop(drop_list, axis=1)
high_correlated_cols(df.drop(drop_list, axis=1), plot=True)
```
